```python
query = d.execute("""SELECT * FROM (SELECT patent_id, name_first, name_last, GROUP_CONCAT(organization) 
                   FROM rawassignee 
                   WHERE patent_id IN {}
                   GROUP BY patent_id) AS sample
                   INNER JOIN patent
                   ON patent.id = sample.patent_id""".format(tuple(corp_id)))

print(d.fetchmany(20))

cols = [column[0] for column in query.description]
results= pd.DataFrame.from_records(data = query.fetchall(), columns = cols)
```

```python
len(results)
```

```python
results.to_csv("patent_org_not.csv", sep="\t")
```
