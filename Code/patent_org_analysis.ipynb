```python
# Company_dict with no repeat or nan values
import math

company = {}

def my_isnan(x):
    try:
        return math.isnan(x)
    except TypeError:
        return False

for key in corp_dict.keys():
    x = set(corp_dict[key].values())
    company[key] = {y for y in x if not my_isnan(y)}
```

```python
import numpy as np

df["abstract"]= df["abstract"].replace(np.nan, '', regex=True)
df["title"] = df["title"].replace(np.nan, '', regex=True)
```

```python
df["text"] = df["abstract"] + " " + df["title"]
```

```python
def pre_processing(text):
    # lowercase
    text=text.lower()
    
    # remove tags
    text=re.sub("&lt;/?.*?&gt;"," &lt;&gt; ",text)
    
    # remove special characters and digits
    text=re.sub("(\\d|\\W)+"," ",text)
    
    return text
```

```python
import nltk

def tokenize(text):
    tokens = nltk.word_tokenize(text)
    stems = []
    for word in tokens:  
        #pos(part of speech) tagging: grammatical tagging of word by their category (verb,noun,etc)
        stems.append(nltk.wordnet.WordNetLemmatizer().lemmatize(word))

    return stems
```

### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
vectorizer = TfidfVectorizer(max_df=0.95, min_df=1, stop_words="english")

corpus = list(df["text"].values)
tfidf = vectorizer.fit_transform(corpus)
tfidf_feature_names = vectorizer.get_feature_names()
```

```python
len(tfidf_feature_names)
```

```python
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_topics=10, max_iter=5, learning_method='online', learning_offset=50.,random_state=0).fit(tfidf)
lda
```

```python
def display_topics(model, feature_names, top_word_number):
    for topic_idx, topic in enumerate(model.components_):
        print ("Topic {}:".format(topic_idx))
        print (" ".join([feature_names[i]
                        for i in topic.argsort()[:-top_word_number - 1:-1]]))

top_words_num = 10

display_topics(lda, tfidf_feature_names, top_words_num)
```
